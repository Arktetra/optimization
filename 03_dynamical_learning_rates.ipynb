# Dynamical Learning Rates

Some strategies for adjusting $\eta$ over time are:

1. piecewise constant
2. exponential decay
3. polynomial decay

```python
import math
import matplotlib.pyplot as plt
```

```python
class ExponentialLR:
    def __init__(self, lr, max_T: int = 10):
        self.lr = lr
        self.t = 0
        self.max_T = max_T

    def __call__(self, lmda = 0.1):
        if self.t >= self.max_T:
            return
        self.t += 1
        return self.lr * math.exp(-lmda * self.t)
    
```

```python
exp_lr = ExponentialLR(lr = 1)
plt.plot([x for x in range(100)], [exp_lr(lmda = 0.1) for _ in range(100)])
```

```python
class PolynomialLR:
    def __init__(self, lr, max_T: int = 10):
        self.lr = lr
        self.max_T = max_T
        self.t = 0

    def __call__(self, alpha = 0.5, beta = 0.9):
        if self.t >= self.max_T:
            return

        self.t += 1

        return self.lr * (beta * self.t + 1) ** (-alpha)
```

```python
poly_lr = PolynomialLR(lr = 0.1)
plt.plot(
    [x for x in range(100)],
    [poly_lr() for _ in range(100)])
```
