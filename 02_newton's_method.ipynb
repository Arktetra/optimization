# Newton's Method

```python
import torch
import matplotlib.pyplot as plt
```

```python
a = torch.tensor(0.5)

def fn(x):
    return torch.cosh(a * x)

def grad_fn(x):
    return a * torch.sinh(a * x)

def hess_fn(x):
    return a ** 2 * torch.cosh(a * x)

def newton(eta = 0.25, steps = 10, start = 10):
    results = [start]
    
    for i in range(steps):
        start -= eta * grad_fn(start) / hess_fn(start)
        results.append(start)
        
    return torch.tensor(results)
```

```python
x = torch.arange(-10., 10., 0.1)
results = newton()
plt.plot(x, fn(x))
plt.plot(results, fn(results), marker = "o")
```
