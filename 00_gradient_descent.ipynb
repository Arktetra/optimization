# Gradient Descent

```python
import matplotlib.pyplot as plt
import numpy as np
```

Let us first define a convex function and its derivative.

```python
def fn(x):
    """A function $f(x) = x^2$."""
    return x ** 2

def dfn(x):
    """Derivative of $f(x)$, $\nabla{f(x)} = 2x$"""
    return 2 * x
```

Plotting the function $f(x) = x^2$.

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, fn(x))
```

Gradient descent is given by:
$$
x \gets x - \eta f'(x)
$$

```python
def gradient_descent(eta, dfn, steps = 10, start = 10):
    results = [start]
    
    for i in range(steps):
        start -= eta * dfn(start)
        results.append(start)
        
    return np.array(results)

results = gradient_descent(eta = .25, dfn = dfn)
plt.plot(x, fn(x))
plt.plot(results, fn(results), marker = "o")
```

With high learning rate:

```python
results = gradient_descent(eta = 0.9, dfn = dfn)
plt.plot(x, fn(x))
plt.plot(results, fn(results), marker = "o")
```

Now, let us condider a non-convex function:
$$
f(x) = xsin(ax)
$$

```python
a = 0.5 * np.pi

def fn(x):
    return x * np.sin(a * x)

def dfn(x):
    return a * x * np.cos(a * x) + np.sin(a * x)
```

```python
results = gradient_descent(eta = 0.5, dfn = dfn, start = 5)
plt.plot(x, fn(x))
plt.plot(results, fn(results), marker = "o")
```

Depending on the choice of the learning rate and how well conditioned the problem is, we may end up with one of many solutions.
